```python
import numpy as np
import pandas as pd 
import random
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
#Taking the  different band images as Input 
image_array = np.ndarray(shape = (512,512,4))

for i in range(4):
    img = plt.imread('band'+ str(i+1)+'.gif')
    image_array[:,:,i] = np.array(img[:,:,0],dtype="float")/np.max(np.array(img[:,:,0]))
    plt.imshow(image_array[:,:,i],cmap='gray')
    plt.subplots_adjust(right=1, top=2)
    plt.show()
    print('band'+ str(i+1)+'.gif')
```

```python
def kmeans(k,image_array):
    r=np.random.randint(0,high=512,size=k)
    c=np.random.randint(0,high=512,size=k)
    D=np.zeros(shape=(512,512,k))
    G=np.zeros(shape=(512,512))
    S=image_array[r,c,:]
    S_new=np.zeros(shape=(k,4))
    variances=0
    
    while(True):
        for i in range(512):
            for j in range(512):
                for c in range(k):
                    D[i,j,c]=np.sqrt(np.sum((S[c,:]-image_array[i,j,:])**2))
                G[i,j]=np.argmin(D[i,j,:]) #Setting a pixel to a cluster
        f=0
        #Updating the means of all clusters
        for i in range(k):
            S_new[i,:]=np.mean(image_array[G==i],axis=0)
            if(np.sqrt(np.sum((S[i,:]-S_new[i,:])**2)))>0.00001:
                f=1
                #print(np.sqrt(np.sum((S[i,:]-S_new[i,:])**2)))

        if f==0:
            break
        else:
            S=S_new
    #To Find Intra-Class Variance
    for i in range(k):
        variances=variances+np.sum((image_array[G==i]-S[i,:])**2)
    
    return variances, G
```

```python
def get_image(k,image_array):
    var_mat=np.zeros(10)
    Clust_mat=np.zeros((512,512,10))
    for i in range(10):
        variances,clusters=kmeans(k,image_array)
        var_mat[i]=variances
        Clust_mat[:,:,i]=clusters
    
    Clust=Clust_mat[:,:,np.argmin(var_mat)]  #Selecting Matrix with minimum Intra-Class Variance
    print(np.min(var_mat))
    Clust=np.round(Clust*(255/(k-1)))
    print(np.unique(Clust))
    return Clust
```

```python
def display(k,image):
    plt.imshow(image,cmap="gray_r") #For Grayscale Image
    plt.subplots_adjust(right=1, top=2)
    plt.title(label=f"Graylevel Image for K = {k}",fontsize=15)
    plt.show()
```

```python
image=get_image(2,image_array)
display(2,image)
```

```python
def show(k,image):
    c=plt.get_cmap('rainbow',k)
    plt.imshow(image,cmap=c)
    plt.subplots_adjust(right=1, top=2)
    plt.title(label=f"Colour-Coded Images for K = {k}",fontsize=15)
    plt.show()
```
